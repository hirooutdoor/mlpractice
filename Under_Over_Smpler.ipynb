```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler
%matplotlib inline
```

```python
data = {
    "n_samples": 200,
    "n_features": 2,
    "n_informative": 2,
    "n_redundant": 0,
    "n_repeated": 0,
    "n_classes": 2,
    "n_clusters_per_class": 1,
    "weights": [0.9, 0.1],
    "random_state": 42
}
```

```python
X, y = make_classification(**data)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], marker="x")
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker="o")
plt.show()
```

```python
n = len(X[y == 0, 0])
```

```python
p = len(X[y == 1, 1])
```

```python
print("Negative", n)
print("Positive", p)
#200件のデータのうち0.9と0.1で分けている
#ここから、アンダーサンプリングで不均衡を解消していく
```

## アンダーサンプリング

```python
sampler = RandomUnderSampler(random_state=42)
```

```python
X_resampled,  y_resampled = sampler.fit_resample(X, y)
```

```python
plt.scatter(X_resampled[y_resampled == 0, 0],X_resampled[y_resampled == 0, 1], marker="x")
plt.scatter(X_resampled[y_resampled == 1, 0], X_resampled[y_resampled == 1, 1], marker="o")
plt.show()
```

```python
n_re = len(X_resampled[y_resampled == 0, 0])
p_re = len(X_resampled[y_resampled == 1, 0])
print("Negative", n_re)
print("Positive", p_re)
#アンダーサンプリングによってpositiveと同じ数の20件になった
```

## オーバーサンプリング

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
data = {
    "n_samples": 200,
    "n_features": 2,
    "n_informative": 2,
    "n_redundant": 0,
    "n_repeated": 0,
    "n_classes": 2,
    "n_clusters_per_class": 1,
    "weights": [0.9, 0.1],
    "random_state": 42
}
X, y = make_classification(**data)
plt.scatter(X[y == 0, 0], X[y == 0, 1], marker="x")
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker="o")
plt.show()
```

```python
sampler_ov = RandomOverSampler(random_state=42)
```

```python
X_resampled,  y_resampled = sampler_ov.fit_resample(X, y)
```

```python
plt.scatter(X_resampled[y_resampled == 0, 0],X_resampled[y_resampled == 0, 1], marker="x")
plt.scatter(X_resampled[y_resampled == 1, 0], X_resampled[y_resampled == 1, 1], marker="o")
plt.show()
```

```python
n_re = len(X_resampled[y_resampled == 0, 0])
p_re = len(X_resampled[y_resampled == 1, 0])
print("Negative", n_re)
print("Positive", p_re)
```
